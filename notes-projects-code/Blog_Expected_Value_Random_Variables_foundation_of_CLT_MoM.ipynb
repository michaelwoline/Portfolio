## Characterising a Distribution using Expected Value



$$\begin{align}
\mu = E[X] & = {} \sum_x x Pr(X=x) \\ \\
Var[X] & = {} E[(x-E[X])^2] \\
    & = {} \sum_x (x-E[X])^2 Pr(X=x)
    \end{align}
$$    


for a binomial distribution
$$\begin{align}
E[X] & ={} \sum_k k \left( \begin{array}{c}n\\k\end{array}\right) p^k (1-p)^{n-k}\\
     & ={} np \\
Var[X] & ={} \sum_k (k-E[X])^2 \left( \begin{array}{c}n\\k\end{array}\right) p^k (1-p^{n-k})\\
    & ={} np(1-p) 
\end{align}
$$


### Expected Values:
- Comes from the model. And, had unknown parameters
- We will almost never know mu, but we can reduce uncertainty of our estimation
- Is a descriptor of the distribution. 
- It is a one number summary of the distribution.
- It can be a measure of the center
- It can be a measure of the spread

- Does not determine the distribution but rather describes something about it. 
    - Example is 2 different distributions with the same $E[X]$
    - One has a wide spread with numbers located equally away from the center, i.e. $E[X]$
    - The 2nd distribution may have values mostly centered around $E[X]$
        - Both have the same $E[X]$, but different distributions.
        

- Don't replace r.v. $X$ with $E[X]$
    - $E[X]$ is a number (if it exists) where as $X$ is a function/distribution 
    - It also ignores the variability of $X$

### A random variable, r.v.s:
   - is a variable that assumes numerical values associated with the random outcomes of an experiment, where one (and only one) numerical value is assigned to each sample point.
    
    
   - Random variables that can assume a countable number of values are called discrete.
   - Random variables that can assume values corresponding to any of the points contained in an interval are called continuous.

- The probability distribution of a discrete random variable is a graph, table, or formula that specifies the probability associated with each possible value that the random variable can assume.

### Proof $Var(X) = E[X^2] - (E[X])^2$

$$
\begin{align}
\sigma^2 = Var[X] & = {} E[X^2] - (E[X])^2 \\
    & = {} \sum_x (x-E[X])^2 Pr(X=x)\\
    & = \sum_x [(x-E[X])(x-E[X])]Pr(X=x)\\    
    & = \sum_x [(x^2 - 2xE[X] + (E[X])^2] Pr(X=x)\\    
    & = \sum_x x^2 Pr(X=x) - \sum_x 2xE[X] Pr(X=x) + \sum_x (E[X])^2 Pr(X=x)\\    
    & = E[X^2] - 2E[X] \sum_x xPr(X=x) + (E[X])^2 \sum_x Pr(X=x)\\    
    & = E[X^2] - 2E[X] E[X] + (E[X])^2 \sum_x 1\\    
    & = E[X^2] - 2(E[X])^2 + (E[X])^2 \\    
    & = E[X^2] - (E[X])^2 \\
    & = E(x^2) - \mu^2, where E(x^2) = \sum_x x^2 Pr(X=x) \\
 \textsf{by rearranging terms we get:} \\
    MoM_2 \rightarrow M_2 & = E[X^2] \\
   E[X^2]   & = \sigma^2 + \mu^2 \\
    E[X^2] - \mu^2 & = \sigma^2 \\
    E[X^2] - (E[X])^2 & = \sigma^2 \\
\end{align}
$$

Let X1, X2, ... , Xn  be a random sample of size n from a distribution (population) with mean μ and variance σ2. What is the mean, that is, the expected value, of the sample mean $\bar X$?

<div style="margin: 30px; padding: 4px; background: none repeat scroll 0pt 50% #99ccff;">
<p><strong>Solution.</strong> Starting with the definition of the sample mean, we have:</p>
<p style="text-align: center;">$$E(\bar{X})=E\left(\dfrac{X_1+X_2+\cdots+X_n}{n}\right)$$ </p>
<p style="text-align: left;">Then, using the linear operator property of expectation, we get:</p>
<p style="text-align: center;">$$E(\bar{X})=\dfrac{1}{n} [E(X_1)+E(X_2)+\cdots+E(X_n)]$$</p>
<p style="text-align: left;">Now, the <em>X<sub>i</sub></em> are identically distributed, which means they have the same mean&nbsp;<em>μ</em>. Therefore,&nbsp;replacing <em>E</em>(<em>X<sub>i</sub></em>) with the alternative notation <em>μ</em>, we get:</p>
<p style="text-align: center;">$$E(\bar{X})=\dfrac{1}{n}[\mu+\mu+\cdots+\mu]$$</p>
<p style="text-align: left;">Now, because there are <em>n</em>&nbsp;<em>μ</em>'s in the above formula, we can rewrite the expected value as:</p>
<p style="text-align: center;">$$E(\bar{X})=\dfrac{1}{n}[n \mu]=\mu $$</p>
<p style="text-align: left;">We have shown that the&nbsp;mean (or expected value, if you prefer) of the sample mean $\bar{X}$ is&nbsp;<em>μ</em>.&nbsp;That is, we have shown that the mean of $\bar{X}$ is the same as the mean of the individual&nbsp;<em>X<sub>i</sub></em>.</p>
</div>

#### Let X1, X2, ... , Xn  be a random sample of size n from a distribution (population) with mean μ and variance σ2. What is the variance of $\bar X$?

<div style="margin: 30px; padding: 4px; background: none repeat scroll 0pt 50% #8AD18A;">
<p><strong>Solution.</strong> Starting with the definition of the sample mean, we have:</p>
<p style="text-align: center;">$$Var(\bar{X})=Var\left(\dfrac{X_1+X_2+\cdots+X_n}{n}\right)$$</p>
<p style="text-align: left;">Rewriting the term on the right so that it is clear that we have a linear combination of <em>X<sub>i</sub></em>'s, we get:</p>
<p style="text-align: center;">$$Var(\bar{X})=Var\left(\dfrac{1}{n}X_1+\dfrac{1}{n}X_2+\cdots+\dfrac{1}{n}X_n\right)$$</p>
<p style="text-align: left;">Then, applying the theorem on the last page, we get:</p>
<p style="text-align: center;">$$Var(\bar{X})=\dfrac{1}{n^2}Var(X_1)+\dfrac{1}{n^2}Var(X_2)+\cdots+\dfrac{1}{n^2}Var(X_n)$$</p>
<p style="text-align: left;">Now, the&nbsp;<em>X<sub>i</sub></em>&nbsp;are identically distributed, which means they have the same variance&nbsp;<em>σ</em><sup>2</sup>. Therefore,&nbsp;replacing <em>Var</em>(<em>X<sub>i</sub></em>) with the alternative notation&nbsp;<em>σ</em><sup style="vertical-align: super; font-size: 0.6em;">2</sup>, we get:</p>
<p style="text-align: center;"><span style="line-height: normal;">&nbsp;$$ Var(\bar{X})=\dfrac{1}{n^2}[\sigma^2+\sigma^2+\cdots+\sigma^2]$$</span></p>
<p style="font: normal normal normal 1em/normal serif; padding-top: 12px; padding-right: 4px; padding-bottom: 12px; padding-left: 4px; line-height: 1.4; margin: 0px;">Now, because there are&nbsp;<em>n</em>&nbsp;<em>σ</em><sup style="vertical-align: super; font-size: 0.6em;">2</sup>'s in the above formula, we can rewrite the expected value as:</p>
<p style="font: normal normal normal 1em/normal serif; padding-top: 12px; padding-right: 4px; padding-bottom: 12px; padding-left: 4px; line-height: 1.4; text-align: center; margin: 0px;">$$Var(\bar{X})=\dfrac{1}{n^2}[n\sigma^2]=\dfrac{\sigma^2}{n}$$</p>
<p style="font: normal normal normal 1em/normal serif; padding-top: 12px; padding-right: 4px; padding-bottom: 12px; padding-left: 4px; line-height: 1.4; margin: 0px;">Our result indicates that as the sample size&nbsp;<em>n&nbsp;</em>increases, the variance of the sample mean decreases. That suggests that on the previous page, if the instructor had taken larger samples of students, she would have seen less variability in the sample means that she was obtaining.&nbsp;This is a good thing, but of course, in general, the costs of research studies no doubt increase as the sample size<em>&nbsp;n&nbsp;</em>increases.&nbsp; There is always a trade-off!</p>
</div>
