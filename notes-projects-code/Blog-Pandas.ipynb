### Interesting behavior of loc vs. iloc vs. row slicing

#### iloc and row slicing behave the same way in that they return rows in the range specified including the first integer to end range n-1.

#### loc behaves differently by returning the full range of rows specified in the range including the end number.
- loc[0:3] returns 4 rows, 0, 1, 2, 3

### How to add data? 
#### What if we want to merge a Series into a Dataframe, making sure the indexes match up.
#### What if we wanted to append some data or quickly merge mulitiple data frames together? 
#### what options are available? 

```python
# a Simple challenge, adding new rows to an existing Dataframe:
# for our example first we will slice up our house_sale_df 
# into new dataframes to manipulate without overwriting the original
temp1_df = house_sale_df.iloc[0:10]
temp2_df = house_sale_df.iloc[10:50]
temp3_df = house_sale_df.iloc[50:150]

appended_df = temp1_df.append(temp2_df, ignore_index= True)
appended_df.tail(3)

# We can ignore the index here because we don't use it to add new rows.
```

```python
#Merge
concat_df = pd.concat([temp1_df, temp3_df, temp2_df])
concat_df.tail(3)
```

```python
#copy off a single column as a series, perform some calculations, and merge back in.
price_series = temp3_df['price']
type(price_series)
```

```python
price_series.tail()
```

```python
# Perform a calculation: add a 5% increase to the price.

price_series = price_series*1.05
price_series.tail()
```

```python
#Now Merge the new series
my_series_merged_with_df = temp3_df.merge(price_series.to_frame(), left_index=True, right_index=True).tail()
my_series_merged_with_df.tail(3)
```

### What just happened?

- we converted the series into a dataframe and retained the index.
- we then merged by making sure the both indexes matched

#### pandas renamed the columns. 
- since we added a 2nd column named price, pandas automatically renamed the first price column to price_x and the new column to price_y

- We can go in and modify these column names.
    - make sure you are passing the dictionary with the new names to the parameter: columns=

```python
my_series_merged_with_df = my_series_merged_with_df.rename(columns = {'price_x':'price', 'price_y': 'price_increase'})
my_series_merged_with_df.tail(3)
```

## How to work with groups:

- Just like SQL, after applying the groupby method, an aggregation function must be applied. 

#### - Apply Changes to groups
#### - Pull out data by groups


```python
#Step 1, group the data based on a column
temp_df = house_sale_df.groupby('bedrooms')
# Step 2, aggregate the data and perform some calculation
temp_df['price'].mean()
```

## groupby based on a calculation:
### suppose we want to group by sales per month per year
### we can either make 2 new columns, 1 for year, 1 for month, and break up the date column and group by the 2 new columns
###  this seems like too much overhead and space, so let's try to do the same with filters.

```python
temp_df = house_sale_df.groupby([house_sale_df['date'].dt.year, house_sale_df['date'].dt.month])
```

```python
print temp_df['price'].mean()
```

### More date groupby options:
[Group By Options:](https://stackoverflow.com/questions/11391969/how-to-group-pandas-dataframe-entries-by-date-in-a-non-unique-column)


### Format the above output for easy reading:



```python
#here is what the format to add commas looks like:
"{:,}".format(100000000)

#Now we will create a lambda function to apply this formatting.
form = lambda x : "{:,}".format(x)

#Now round and apply the format to the mean of the grouped by year and month. 
temp_df['price'].mean().round(0).apply(form)
```

### Use pivot instead of Groupby to accomplish the same task. 

```python
my_piv = (pd.pivot_table(house_sale_df, values = 'price', 
                index = [house_sale_df['date'].dt.year, house_sale_df['date'].dt.month] 
                ).round(0))

my_piv = my_piv['price'].apply(form)
my_piv
```

### Now we change a column to a categorical
### Add the category labels such as 'yes', 'no'
### The data is stores as the category and displayed as label 

```python
house_sale_df.waterfront = pd.Categorical(house_sale_df.waterfront)
house_sale_df.waterfront.cat.categories = ['no','yes']
house_sale_df.waterfront.cat.codes.unique()
```

```python
print house_sale_df['waterfront'].sort_values(ascending=True).head(3)
print house_sale_df['waterfront'].sort_values(ascending=True).tail(3)
```

### Create a table directly to a postgres database and add the data.

- First convert the date column of the pandas dataframe to a datetime type, so the database will recognize it as a date type. 

```python
house_sale_df[['date']] = house_sale_df[['date']].apply(pd.to_datetime)
house_sale_df.loc[:0, 'date']
```

```python
from sqlalchemy import create_engine
import psycopg2
engine = create_engine('postgresql://@localhost:5432/kings_county_sales')

house_sale_df.to_sql("kc_house", engine)
```

### Now read the data from Postgres into a Dataframe

```python
df = pd.read_sql_query('select * from "kc_house"',con=engine)
df.tail(3)
```
