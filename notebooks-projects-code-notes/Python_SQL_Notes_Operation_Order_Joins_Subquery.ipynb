Terms:
* RDBMS
* Keys - used for joins.
* ACID:
    * Atomicity
    * Consistency
    * Isolatin
    * Durability
* Primary key 
    * Uses binary tree search very fast and efficient.
    * Used to do joins etc.
* Foreign Key
    * The key that is referenced in another table. 
    * This key is saved in table 1, so table 1, and reference table 2.
        * Table 2 is called the primary key
        * Table 1, Primary key has foreign key which is used to connects to table 2 that has the primary key.
