* Algorithms used for (Clustering):
    * K-means
        * find where the centroids are for the data. 
            1. drop some centroid points into the data space
            2. figure out using say L1 or L2 distance which points should be assigned to a centroid.
            3. Then average the locations and move the centriod to that middle average of the points that were assigned.
            4. move the centroid to the new closest average of the points assigned.
            5. Repeat steps 2, 3, and 4 above
                1. until the model becomes stable.
                2. This can get stuck in the local minimum, so choose multiple different starting centriod points and score to see which ones wind up the most staple and consistent locations. 
                3. Different starting values can cause it to go to different end result centroid groupings. 



    * hierachical clustering
        1. start by finding the 2 closest points within the data space. 
        2. Once we have the first 2 we find one of the optional distances
            1. Options: (each choice has ramifications to changing the outcome)
                1. closest points
                2. farthest points
                3. centroid (use the central point)
                4. average (keep points but use each one to calculate average from others)
            2. then we find the next 2 closest points by comparing all points to each other and the grouped points optional measure point from above. 
            3. draw the distance in height 
            4. we will have 2 heights now 1 for the first 2 points, and 1 for the comparison point. 
            5. then find the next 2 closest points. (repeat process)
            6. Dendrogram is a height height measure graph tool. It looks like a tree diagram. 
                1. these don’t really tell us anything because it based on our 4 optional distance choices. so they don’t have much meaning
                2. at some point there are so many data points we will not be able to draw or read the diagram. 



    * mixture models
        *  along 1 feature set it on a flat line
        * create a kde across the data.
        * apply a couple of normal distribution curves.
            * these curves should look in combination close to the KDE curvatures
        * K is how many simple curve normal distributions to combine
            * Can use EM-algorithm
            * Bayesian Mixture Models



Choosing K is the hardest part of the process:


How to choose K:
1. Elbow Method
    * diminishing returns algorithms.
    * Run the algorithm from above, k-means, hierarchical clustering, mixture models.
    * sum up the distances from the centroid point to each data point,  perform this for each centroid k point. 
    * when we see an elbow, we have a clear point of where to set k. 
    * many times it will look smoother, rounder, and not clear at what number to set our k

* Silhouhette 
    * notion: merging close clusters.
    * measures separation between current cluster and next best cluster sort of like finding a margin of some kind. 
    * 0.take a point in a cluster, and measure average distance to all points.
    1. then go through each point in the cluster, and take the average distance to all other points within the same cluster.
    2. 2nd set of measures with every point of the cluster, take the average distance of the next closest clusters points.  
    3. calculate the score:
        * take the 2ndary score - primary score / secondary score
        * score will be between 0 and 1
        * if close to 1, then average distance in primary cluster is way better than next best cluster.
        * if close to 0, then I don’t have a strong preference for the cluster that I am in. distance in current cluster is almost the same as the average distance to secondary cluster. 
    1. 4th step: silhouette score:
        * between 0 and 1
        * add up all of the scores from step 3
        * cycle through all of the k’s for the algorithm. 
        * go through the scoring process
        * find the score that maximizes the overall score
            * higher the score the better segregated and separated the clusters are from each other. 
        *  rule of thumb, higher than .8 

* gap statistic
    * Tiborishini of statistical learning
    * works all of the time for simpler problems 
    * take the clustered data on the left and measure distnaces from the centroid
    * randomize another set of data and take distances from centroid
    * Add more centroids for both and again measure distances 
    * take the difference between the random distance mse and clustered data.
    * score by:
        * when cluster point k value is greater than the lower bound of the random data's variance of the next k
    * process uses an incremental hypothesis test until the there is a failure between the k we are on and the next k to test against. 
    * when we fail to reject the null hypothesis of the next k, we stop and use the current k we are testing
    * H null - no significant improvement with k+1 
    * when we fail to reject we stop
    * when we reject the null hypothesis, we keep moving because I reject that there is no signifivanct improvement with k+1

Principal Component Analysis: 
* Produces a low-dimensional representation of a dataset.
* Finds sequence of linear combinations of variables having maximal variance and mutually uncorrelated
* Derives variables for use in supervised learning.
* Tool for data visualization also…

* A way of reducing down to the most important features.
* Based on largest variance.

* Plot data points against a linear line. In example there are only 2p.
